```python
ClassData['Coops/Internships (YN)']=ClassData['Coops/Internships (YN)'].str.upper()
coops=ClassData['Coops/Internships (YN)']
coopyes=numpy.count_nonzero(coops=='Y')
print(coopyes)
coopno=numpy.count_nonzero(coops=='N')
print(coopno)
totalcoops=coopyes+coopno
yespercent=coopyes/totalcoops
nopercent=coopno/totalcoops
print(yespercent,nopercent)

def percentage(a, b):
    return round(a / b * 100, 2)
yesper=percentage(coopyes,totalcoops)
noper=percentage(coopno,totalcoops)
print(yesper,noper)
```
